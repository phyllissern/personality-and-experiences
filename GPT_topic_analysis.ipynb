```python
#### RUN THIS: FINAL EVALUATION METHOD ####
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances
import scipy.spatial as sp
import scipy.sparse as sps
# from scipy.spatial.distance import jensenshannon
from scipy.stats import wasserstein_distance, spearmanr
from time import time
from skbio.stats.distance import mantel
from sklearn.decomposition import NMF

def _special_sparse_dot(W, H, X):
    """Computes np.dot(W, H), only where X is non zero."""
    if sps.issparse(X):
        ii, jj = X.nonzero()
        n_vals = ii.shape[0]
        dot_vals = np.empty(n_vals)
        n_components = W.shape[1]

        batch_size = max(n_components, n_vals // n_components)
        for start in range(0, n_vals, batch_size):
            batch = slice(start, start + batch_size)
            dot_vals[batch] = np.multiply(W[ii[batch], :], H.T[jj[batch], :]).sum(
                axis=1
            )

        WH = sps.coo_matrix((dot_vals, (ii, jj)), shape=X.shape)
        return WH.tocsr()
    else:
        return np.dot(W, H)

def beta_divergence(X, W, H, beta, square_root=False):
    """
    X : float or array-like of shape (n_samples, n_features)

    W : float or array-like of shape (n_samples, n_components)

    H : float or array-like of shape (n_components, n_features)
    """
        
    EPSILON = np.finfo(np.float32).eps

    if sps.issparse(X):
        # compute np.dot(W, H) only where X is nonzero
        WH_data = _special_sparse_dot(W, H, X).data
        X_data = X.data
    else:
        WH = np.dot(W, H)
        WH_data = WH.ravel()
        X_data = X.ravel()
    
    # do not affect the zeros: here 0 ** (-1) = 0 and not infinity
    indices = X_data > EPSILON
    WH_data = WH_data[indices]
    X_data = X_data[indices]

    # used to avoid division by zero
    WH_data[WH_data < EPSILON] = EPSILON

    # generalized Kullback-Leibler divergence
    if beta == 'kullback-leibler':
        # fast and memory efficient computation of np.sum(np.dot(W, H))
        sum_WH = np.dot(np.sum(W, axis=0), np.sum(H, axis=1))
        # computes np.sum(X * log(X / WH)) only where X is nonzero
        div = X_data / WH_data
        res = np.dot(X_data, np.log(div))
        # add full np.sum(np.dot(W, H)) - np.sum(X)
        res += sum_WH - X_data.sum()

    if square_root:
        res = max(res, 0)  # avoid negative number due to rounding errors
        return np.sqrt(2 * res)
    else:
        return res

t0 = time()
data_matrix = bow_matrix

# theoretical distance matrix to compare subset data to
theoretical_similarity_matrix = theoretical_similarity_matrix                           # matrix to compare to; theoretical
theoretical_distance_matrix = 1-theoretical_similarity_matrix                           

# empirical distance matrix to compare to
empirical_distance_matrix = original_distance_full
empirical_similarity_matrix = original_similarity_full


# Initialize lists to store eval metrics
wasserstein_subset = []
wasserstein_all = []
mantel_theoretical_subset = []
mantel_empirical_subset = []
mantel_empirical_all = []
reconstruction_errors_subset = []
reconstruction_errors_all = []

# Fixed NMF parameters
init = "nndsvda"
n_top_words = 5
beta_loss_func = "kullback-leibler"

# Iterate over different numbers of topics
for num_topics in test_range:
    
    # Fit NMF model
    nmf_model = NMF(
                    n_components=num_topics,
                    random_state=42,
                    init=init,
                    beta_loss=beta_loss_func,
                    solver="mu",
                    alpha_W=0.00005,
                    alpha_H=0.00005,
                    l1_ratio=0.5,
                    max_iter=1000,
                    ).fit(data_matrix)
    
    nmf_topic_matrix = nmf_model.transform(data_matrix)

    ### RECONSTRUCTION ERROR
    # on subset- manually computed
    subset_reconst_errs = beta_divergence(X=data_matrix[matching_indices], 
                                          W=nmf_topic_matrix[matching_indices], 
                                          H=nmf_model.components_, 
                                          beta='kullback-leibler', 
                                          square_root=False)
    reconstruction_errors_subset.append(subset_reconst_errs)
    reconstruction_errors_all.append(nmf_model.reconstruction_err_)     # on full data

    ### WASSERSTEIN ON PROBABILITY DISTRIBUTIONS (NOT DISTANCE MATRICES)    
    wasserstein_dist_full = np.mean(np.array([wasserstein_distance(u, v) for u, v in zip(tfidf_matrix.toarray(), nmf_topic_matrix)]))
    wasserstein_dist_subset = np.mean(np.array([wasserstein_distance(u, v) for u, v in zip(tfidf_matrix.toarray()[matching_indices], nmf_topic_matrix[matching_indices])]))    
    wasserstein_subset.append(wasserstein_dist_full)
    wasserstein_all.append(wasserstein_dist_subset)

    ### MANTEL TEST (SPEARMAN CORRELATION): compare to theoretical AND empirical
    # get distance matrix with nmf topic vectors
    nmf_distance_matrix = cosine_distances(nmf_topic_matrix)
    nmf_distance_subset = nmf_distance_matrix[matching_indices][:,matching_indices]

    mantel_theoretical_subset.append(mantel(nmf_distance_subset, theoretical_distance_matrix, method='spearman'))
    mantel_empirical_subset.append(mantel(nmf_distance_subset, empirical_distance_matrix[matching_indices][:,matching_indices], method='spearman'))
    mantel_empirical_all.append(mantel(nmf_distance_matrix, empirical_distance_matrix, method='spearman'))


print("done in %0.3fs." % (time() - t0))
```
